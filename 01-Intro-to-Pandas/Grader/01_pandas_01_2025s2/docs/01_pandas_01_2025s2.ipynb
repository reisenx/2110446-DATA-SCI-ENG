# Problem `Q1`

Please find the shape of the data.

Shape of the `DataFrame` is a tuple with two values.
- The first one is **amount of rows**.
- The second one is **amount of columns**.

To access shape of the `DataFrame`, use `.shape`

```python
scores_df_shape = SCORES_DF.shape
print(f"Shape of the DataFrame is {scores_df_shape}.")
```

- Use `.shape[0]` for **amount of rows**.
- Use `.shape[1]` for **amount of columns**.

# Problem `Q2`

Please find the max score of the data.

The `DataFrame` has two columns.
- **`id` column** is student ID.
- **`score` column** is student score.

**NOTE:** `.head(10)` is a command to only show the first 10 rows.

```python
SCORES_DF.head(10)
```

To access each column use `[]` operator.
- Use `SCORES_DF["id"]` to access `id` column.
- Use `SCORES_DF["score"]` to access `score` column.

```python
SCORES_DF["id"].head(5)
```

```python
SCORES_DF["score"].head(5)
```

`[]` operator can also access multiple columns.
- Use `SCORES_DF[["score", "id"]]` to access `score` and `id` column in order.

```python
SCORES_DF[["score", "id"]].head(5)
```

To find maximum value of the column, use `.max()`.

```python
max_score = SCORES_DF["score"].max()
print(f"The maximum score is {max_score} points.")
```

# Problem `Q3`

Please find the total student that have score equal or more than 80 points.

- To access score column, we use `SCORE_DF["score"]`.
- So the condition score more than or equal to 80 point would be `SCORE_DF["score"] >= 80`
- To filter rows that meets the condition, use `[]` operation.

```python
# Create filter condition.
CONDITION = SCORES_DF["score"] >= 80

# Apply filtering to the DataFrame.
SCORES_DF[CONDITION]
```

To find amount of rows, use `.shape[0]`

```python
dataset_rows = SCORES_DF[CONDITION].shape[0]
print(f"There are {dataset_rows} students who has more than or equal to 80 points.")
```
